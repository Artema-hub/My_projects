# Исследование объявлений о продаже квартир

Получены данные сервиса недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Задача — установить параметры. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

### Шаг 2. Предобработка данных

```python
# Определяем количество дубликатов строк
df.duplicated().sum()
```

```python
# Определяем количество пропущенных значений в %
df.isna().mean()
```

```python
df.describe()
```

Столбец 'total_images'.Пропусков нет, тип данных int, что полностью соответствует логике.

Столбец 'last_price'. Пропусков нет, тип float. Заменим тип на int, что облегчит данные и не приведет к потере данных. 

```python
# Замена типа данных float на int в столбце 'last_price'
df['last_price'] = df['last_price'].astype(int)

# Проверка
df.info()
```

Столбец 'total_area'. Пропусков нет, тип данных float, что полностью соответствует логике.

Столбец 'first_day_exposition'. Пропусков нет, тип object. Переведем дату размещения объявления в тип data.

```python
# Приводим столбец 'first_day_exposition' к формату date
df['first_day_exposition'] =  pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d')

df.head()
```

Столбец 'rooms'. Пропусков нет, тип данных int, что полностью соответствует логике. В столбце присутствует 194 нулевых значения, чего не может быть. Удалим эти строки.

```python
# Считаем количество строк с нулевым количеством комнат
df[df['rooms'] == 0]['rooms'].count()
```

```python
# Удаляем строки со значением комнат нуль
df = df[df['rooms'] != 0]

# Проверка
df[df['rooms'] == 0]['rooms'].count()
```

Столбец 'ceiling_height'. Пропусков 39%. Все пропуски заменим медианой. Медиану будем считать только по тем строкам, где высота потолков лежит в диапазоне от 2.2м до 3.2. Из диаграммы размаха видим, что высоту потолка вне указанного диапазона можно считать выбрасом. Тип данных float, что полностью соответствует логике.

```python
# Диаграмму размаха столбца 'ceiling_height'
plt.ylim(0, 5)
df.boxplot(column='ceiling_height')
```

```python
# Считаем медиану столбца 'ceiling_height' для значений 'ceiling_height' в диапазоне (2.2, 3.2)
df_celing_hight_main = df[(df['ceiling_height'] > 2.2)&(df['ceiling_height'] < 3.2)]

celing_hight_main_median = df_celing_hight_main['ceiling_height'].median()
celing_hight_main_median
```

```python
# Заменяем пропуски в столбце 'ceiling_height' медианой celing_hight_main_median.
df['ceiling_height'] = df['ceiling_height'].fillna(celing_hight_main_median)

# Проверка
df.isna().mean()
```

Столбец 'floors_total'. Количество этажей всегда целое число, поэтому тип float заменим на int. Количество пропусков 0.36%, что очень мало и их удаление не приведет к искажению результатов исследования.

```python
# Просматриваем строки с пропусками в столбце 'floors_total'.
df[df['floors_total'].isna()].head(10)
```

```python
# Удаляем пропуски в столбце 'floors_total'
df.dropna(subset = ['floors_total'], inplace = True)

df.isna().mean()
```

```python
# Заменяем тип float на int в столбце 'floors_total'.
df['floors_total'] = df['floors_total'].astype(int)

# Проверка
df.info()
```

Столбец 'living_area'. Имеет тип float, что полностью соответсвует логике данных. Количество пропусков 8%. Скорее всего дающие объявление просто не знали и не заполнили данную графу. Заполним пропуски значением общей площади 'total_area' умноженной на медиану доли 'living_area' в total_area.

```python
# Смотрим строки с пропусками в столбце 'living_area' 
df[df['living_area'].isna()].head()
```

```python
# Считаем медиану доли 'living_area' в 'total_area'
living_total_share_median = (df['living_area'] / df['total_area']).median()

print(living_total_share_median)

# Заполним пропуски в столбце 'living_area' значением общей площади 'total_area'
# умноженной на медиану доли 'living_area' в total_area.
df['living_area'] = df['living_area'].fillna(df['total_area']*living_total_share_median)

# Проверяем
df.head(7)
```

Столбец 'floor'. Пропусков нет, тип данных int, что полностью соответствует логике.

Столбец 'is_apartment'. Пропусков 88%. Скорее всего человек не понял значения этого поля и не заполнил его. Значений True (нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания) всего 50. Поскольку мы исследуем объявления о продаже квартир, то очевидно, что почти все они должны иметь значение False в этом столбце. Заменим пропуски на False и тип object на bool.

```python
# Заменяем пропуски в столбце 'is_apartment' на False
df['is_apartment'] = df['is_apartment'].fillna(False)

# Заменяем тип object на bool в столбце 'is_apartment'
df['is_apartment'] = df['is_apartment'].astype(bool)

# Проверяем
df.isna().mean()
```

Столбец 'studio'. Пропусков нет. Тип bool, что полностью соответствует логике данных.

Столбец 'open_plan'. Пропусков нет. Тип bool, что полностью соответствует логике данных.

Столбец 'kitchen_area'. Пропусков 9%. Скорее всего люди дающие объявление просто не заполнили это поле, скорее всего потому, что не знают. Заполним пропуски значением общей площади 'total_area' умноженной на медиану доли 'kitchen_area' / 'total_area'. Столбец имеет тип float, что соответствует логике данных.

```python
# Считаем медиану доли 'kitchen_area' / 'total_area'
kitchen_total_share_median = (df['kitchen_area'] / df['total_area']).median()

print(kitchen_total_share_median)

# Заполним пропуски в столбце 'living_area' значением общей площади 'total_area'
# умноженной на медиану доли 'living_area' в total_area.
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area']*kitchen_total_share_median)

# Проверяем
df.head()
```

Столбец 'balcony'. Пропусков 47%. Целесообразно предположить, что отсутствие данных в этом столбце означает отсутствие балкона, и что поэтому продавец просто не заполнил это поле. Заменим пропуски на нуль. Столбец имеет тип float, который заменим на int, т.к. кол-во балконов это всегда целое число.

```python
# Заменяем пропуски в 'balcony' на нуль
df['balcony'] = df['balcony'].fillna(0)

# Меняем тип столбца 'balcony' с float на int
df['balcony'] = df['balcony'].astype(int)

# Проверяем отсутствие пропусков в 'balcony'
df.isna().mean()
```

Столбец 'locality_name'. Пропусков 0.2%. Пропусков слишком мало чтобы удаление этих строк оказало хоть сколь-нибудь значимое влияние на наше исследование. Пропуски удалим. Столбец имеет тип данных object, что соответствует логике данных. 

```python
# Удаляем строки с пропусками в столбце 'locality_name'.
df.dropna(subset=['locality_name'], inplace=True)

df.isna().mean()
```

```python
# Первые 10 уникальных значений столбца'locality_name'
df['locality_name'].value_counts().head(10)
```

Столбецы 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000, 'ponds_nearest', 'days_exposition'. Пропуски заменим медианным значением, т.к. нет возможности более точного определения значений. Столбецы имеют тип float, что соответствует смыслу данных.

```python
# Заменяем пропуски в столбце 'airports_nearest' медианой
df['airports_nearest'] = df['airports_nearest'].fillna(df['airports_nearest'].median())

# Заменяем пропуски в столбце 'cityCenters_nearest' медианой
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(df['cityCenters_nearest'].median())

# Заменяем пропуски в столбце 'parks_around3000' медианой
df['parks_around3000'] = df['parks_around3000'].fillna(df['parks_around3000'].median())

# Заменяем пропуски в столбце 'parks_nearest' медианой
df['parks_nearest'] = df['parks_nearest'].fillna(df['parks_nearest'].median())

# Заменяем пропуски в столбце 'ponds_around3000' медианой
df['ponds_around3000'] = df['ponds_around3000'].fillna(df['ponds_around3000'].median())

# Заменяем пропуски в столбце 'ponds_nearest' медианой
df['ponds_nearest'] = df['ponds_nearest'].fillna(df['ponds_nearest'].median())

# Заменяем пропуски в столбце 'days_exposition' медианой
df['days_exposition'] = df['days_exposition'].fillna(df['days_exposition'].median())

# Заменяем тип 'days_exposition' с float на int
df['days_exposition'] = df['days_exposition'].astype(int)

# Проверяем
df.isna().mean()
```

```python
df.info()
```

### Вывод предобработка

Исходные данные составляли 23696 строки. После предобработки осталось 23371 строка. Были изменены типы данных некоторых столбцов исходя из характера хранящихся в них данных.

### Шаг 3. Посчитаем и добавим в таблицу расчетные данные

```python
# Добавляем столбец 'price_m' со стоимостью одного метра общей площади квартиры
df['price_m'] = (df['last_price'] / df['total_area']).astype(int)

df.info()
```

```python
# Добавляем столбец с днем недели 'week_day' в который было размещено объявление
df['week_day'] = df['first_day_exposition'].dt.weekday_name

df.head()
```

```python
# Добавляем столбец 'mounth' c месяцем и годом в который было размещено объявление
df['month'] = df['first_day_exposition'].dt.month

df.head()
```

```python
# Добавляем столбец 'year' c месяцем и годом в который было размещено объявление
df['year'] = df['first_day_exposition'].dt.year

df.head()
```

```python
# Категоризация этажа квартиры
df['floor_category'] = (df['floor'].where(~(df['floor'] == df['floors_total']), 'последний')
                       .where(~(df['floor'] == 1), 'первый')
                       .where(((df['floor'] == 1) | (df['floor'] == df['floors_total'])), 'другой')
)

# Проверка
df[['floor_category', 'floor', 'floors_total']].head(15)
```

```python
# Добавляем столбец 'living_total_share' с отношением living_area к total_area
df['living_total_share'] = df['living_area'] / df['total_area']

df.head()
```

```python
# Добавляем столбец 'kitchen_total_share' c отношением kitchen_area к total_area
df['kitchen_total_share'] = df['kitchen_area'] / df['total_area']

df.head()
```

* 'year_mounth' c месяцем и годом в который было размещено объявление

* 'floor_category' - категоризация этажа квартиры на: 'первый', 'последний', 'другой'

* 'living_total_share' - отношение 'living_area' к 'total_area'

* 'kitchen_total_share' - отношение 'kitchen_area' к 'total_area'

Диаграммы показывают, что квартиры начинают нормально продаваться уже с первых дней объявлений, а вот продажа более чем за 600 дней уже считается аномально долгой.

```python
# Квартиры с ценой до 10000000

df_limit_price = df.query('last_price < 10000000')
```

```python
# Точечная диаграмма 'last_price' от 'total_area'
df_limit_price.plot(x='last_price', y='total_area', kind='scatter', figsize=(10, 6), alpha=0.2,)
plt.show()

# К-т корелляции
df_limit_price['last_price'].corr(df_limit_price['total_area'])
```

#### Таким образом, стоимость квартир снижается при снижении их площади. Но стоит отметить, что зависимость эта слабая, о чем говорит к-т корреляции равный 0,65, и всегда можно найти квартиру, которая стоит дешевле, но имеет большую площадь.

```python
# Диаграмма boxplot для 'last_price' от 'rooms'
df_limit_price.boxplot(column = 'last_price', by = 'rooms')
plt.show()

# К-т корелляции
df_limit_price['last_price'].corr(df_limit_price['rooms'])
```

#### Таким образом стоимость квартир растет с увеличением количества комнат. Но стоит отметить, что зависимость эта не сильная, о чем говорит к-т корреляции равный 0,42 и часто можно найти квартиру, которая стоит дешевле, но комнат в ней больше. 

```python
# Точечная диаграмма 'price_m' от 'cityCenters_nearest'
df_limit_price.plot(x='price_m', y='cityCenters_nearest', kind='scatter', figsize=(10, 6), alpha=0.3,)
plt.show()

# К-т корелляции
df_limit_price['price_m'].corr(df_limit_price['cityCenters_nearest'])
```

#### Зависимость стоимости кв.м. квартиры от ее удаленности имеет очень слабый отрицательный характер и практически отсутствует, о чем говорит к-т корреляции равный -0,18.

```python
# Boxplot для 'price_m' от 'floor_category'
df_limit_price.boxplot(column = 'price_m', by = 'floor_category')
plt.show()
```

#### Из диаграммы выше видим зависимость цены квртиры между категориями этажа. Так цены на квартиры, с категорией этажа 'первый' и 'последний' в среднем имеют цену цуть ниже чем 'другой'. Но стоит отметить, что эта зависимость не сильная.
